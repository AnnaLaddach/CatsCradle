```python
import scanpy as sc
import pandas as pd
```

```python
#read data
adata = sc.read("adata_anno.h5ad")
```

```python
#scale data
adata.layers['scaled'] = sc.pp.scale(adata, copy=True).X
```

```python
#extract scaled matrix for highly variable genes
data_mat = adata.layers["scaled"][:,adata.var["highly_variable"]].T
dat = pd.DataFrame(data_mat)
dat.index = adata.var_names[adata.var["highly_variable"]]
dat.columns = adata.obs_names
dat.to_csv("genes_hvar_scaled.csv")
```

```python
#read this in as an anndata object where coloums are genes and rows are cells
adata_genes = sc.read("genes_hvar_scaled.csv")
```

```python
#compute PCA
sc.tl.pca(adata_genes, svd_solver='arpack')
sc.pl.pca_variance_ratio(adata_genes)
```

```python
#compute UMAP and clustering
sc.pp.neighbors(adata_genes, n_pcs = 10)
sc.tl.umap(adata_genes)
sc.tl.louvain(adata_genes, key_added='louvain_r1')
sc.pl.umap(adata_genes, color = ['louvain_r1'])
```
